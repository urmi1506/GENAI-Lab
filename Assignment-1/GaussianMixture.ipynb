```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
```

```python
data1= np.random.normal(40,5,100)
data2 = np.random.normal(70,5,100)

data = np.concatenate([data1,data2])

data = data.reshape(-1,1)
```

```python
gmm = GaussianMixture(n_components=2,random_state=0)
gmm.fit(data)
```

```python
lables = gmm.predict(data)

probablities=gmm.predict_proba(data)

print("first 5 data points probablities:")
print(probablities[:5])
```

```python
print("Means of Gaussians:", gmm.means_)
print("Variances of Gaussians:", gmm.covariances_)
print("Weights of Gaussians:", gmm.weights_)
```

```python
plt.scatter(data, np.zeros_like(data), c=lables, cmap='viridis')
plt.title("Gaussian Mixture Model Clustering")
plt.xlabel("Data Values")
plt.show()
```
